```python
# Load libraries

import xarray as xr
import matplotlib.pyplot as plt
```

# North Pole Perturbations

## Testing Input

```python
# Testing input of north pole perturbation

test_northpole = xr.open_dataset('/data/keeling/a/mailes2/a/CESM_DATA/aqua_northpoleperturb/outputdata/atm/hist/aqua_northpoleperturb.cam.h0.0001-01.nc')
test_control = xr.open_dataset('/data/keeling/a/pappup2/a/CESM/cam5_new/CESM_output_data/atm/hist/cam5_new.cam.h0.0001-01.nc')
```

```python
test_northpole.mean('lon').TS.plot(label='North Pole Perturbation')
test_control.mean('lon').TS.plot(label='Control')
plt.title('Sea Surface Temperature Test')
plt.legend()
```

```python
test_deltaNP = test_northpole.mean('lon').TS - test_control.mean('lon').TS
test_deltaNP.plot()
plt.title('Delta SST for the North Pole')
```
