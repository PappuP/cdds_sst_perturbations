# Control Sea Surface Temperature Plotting - Python

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Range of latitudes
lats = np.linspace(-np.pi/2, np.pi/2, 50)
```

```python
# Setting SST to Control
SST = np.zeros(50)
i = 0
for lat in lats:
    if lat < -np.pi/3:
        SST[i] = 0
    elif lat > np.pi/3:
        SST[i] = 0
    else:
        SST[i] = 27 * (1 - np.sin(3*lat/2)**2)
    i += 1
```

```python
#Plotting Control SST
lats_degrees = lats * 180/np.pi
plt.plot(lats_degrees, SST)
plt.title("Control SST")
plt.xlabel("Latitude")
plt.ylabel("SST (C)")
```
